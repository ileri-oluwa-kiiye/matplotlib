# Distributions

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
iris_filepath = './data/iris.csv'

iris_data = pd.read_csv(iris_filepath, index_col='Id')

iris_data
```

# Histograms
Let's create a histogram to see how petal length varies in iris flowers.

```python
sns.histplot(iris_data['Petal Length (cm)'], kde=True)
```

**KDE(Kernel Desnsity Estimate)** is a way to estimate the probability density function (PDF) of a continuous random variable. In simpler terms, it provides a smoothed curve that represents the distribution of the data. It’s particularly useful for visualizing the distribution of data points in a way that highlights its shape, spread, and central tendency.

We can also make a KDE plot that shows that kde line without the histogram.

```python
sns.kdeplot(iris_data['Petal Length (cm)'], shade=True)
```

### Histogram for Petal Lenth by species

```python
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')
```

```python
#We can make a kde plot for this too

sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', shade=True)
```
