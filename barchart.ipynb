# Bar Charts and Heatmaps
They are so similar to the linecharts, it's virtually the same thing

```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline

pd.plotting.register_matplotlib_converters()
import seaborn as sns

print("Setup complete")
```

### Load your dataset

```python
flight_filepath = './data/flight_delays.csv'

flight_data = pd.read_csv(flight_filepath, index_col='Month')

flight_data
```

```python
plt.figure(figsize=(10,6))

plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

sns.barplot(data=flight_data, x=flight_data.index, y=flight_data['NK'])
```

# Heatmaps

Heatmaps are a quick way to visualize data, using color codes.

```python
plt.figure(figsize=(16,6))

plt.title("Average Arrival Delay for Each Airline, by Month")

sns.heatmap(data=flight_data, annot=True)
#The annot is to show the values in the heatmap
```

With this heatmap, you can easily get an overview of your datset
