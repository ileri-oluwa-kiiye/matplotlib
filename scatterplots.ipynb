# Scatterplots

These are used to display relationships between variables.

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
insurance_filepath = "./data/insurance.csv"

insurance_data = pd.read_csv(insurance_filepath)

insurance_data.head()
```

To use a scatterplot, we typically have to use the sns.scatterplot command and set values for;
- The x axis and
- The y axis
as shown below

```python
plt.figure(figsize=(10,6))

sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

This scatterplot above shows the relationship between the insurance charges and the customer's BMI. From the figure, we can see that they are positively correlated.

You can use the sns.regplot command to add a regression line to the scatterplot as shown below;

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

The line is positively sloped. i.e. the greater the customer's bmi, the greater the insurance charges usually.

## Color coded scatterplots

Instead of using scatterplots to display the relationship between two variables, we can use it to display the relationship between more than two variables, using color coded scatterplots.



```python
sns.scatterplot(data=insurance_data, x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

Using a hue that's not a binary categorical data;

```python
sns.scatterplot(data=insurance_data, x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['region'])
```

To add more than one regression line i.e. in a case where we're using the hue, we use the command sns.lmplot. Here, we have to just set the x,y and hue values to the name of the column only i.e. 'charges' instead of insurance_data['charges]

```python
sns.lmplot(x="bmi", y="charges", hue="sex", data=insurance_data)
```
